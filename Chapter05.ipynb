```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import yfinance as yf
```

```python
symbol = 'AAPL'
df = yf.download(symbol, start='2022-01-01', end='2023-01-01')
df.index = pd.to_datetime(df.index)
df
```

```python
window = 3
SMA1 = 'SMA-' + str(window)
df[SMA1] = df['Close'].rolling(window).mean()
col_names = ['Close', SMA1]
df2 = df[col_names]
df2
```

```python
# plt.figure(figsize=(15, 7))
colors = ['blue', 'red']
# df['Close'].plot()
df2.plot(color=colors, linewidth=3, figsize=(12, 6))
plt.xticks(fontsize=13)                
plt.yticks(fontsize=13)
col_names = ['Close', SMA1]
plt.legend(labels = col_names, fontsize=13)                
# title and labels                
plt.title('Daily adjusted closing price and its SWA', fontsize=20)                
plt.xlabel('Date', fontsize=16)                
plt.ylabel('Price', fontsize=16)
```

```python
window3 = 20
SMA2 = 'SMA-'+str(window3)
df3 = df2.copy(deep=True)
df3['SMA-' + SMA2] = df2['Close'].rolling(window3).mean()
col_names3 = ['Close', SMA1, SMA2]

colors = ['blue', 'red', 'green']
df3.plot(color=colors, linewidth=3, figsize=(20, 10))

plt.xticks(fontsize=13)
plt.yticks(fontsize=13)
plt.legend(labels=col_names3, fontsize=13)

plt.title('Daily adjusted closing price and its SWA', fontsize=20)
plt.xlabel('Date', fontsize=16)
plt.ylabel('Price', fontsize=16)
```
